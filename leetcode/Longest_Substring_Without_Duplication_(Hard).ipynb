# Longest Substring Without Duplication

Write a function that takes in a string and returns its longest substring without duplicate characters.

You can assume that there will only be one longest substring without duplication.

Sample Input

```
string = "clementisacap"
```



Sample Output

```
"mentisac"
```



# Solution 1:

move left idx 1 by 1 when found duplicate

````python
# Time : O(n)
# Space : O(min(n, c))
# n = length of string
# a = no. of distinct char
def longestSubstringWithoutDuplication(string):
  left_idx = 0
  right_idx = 0
  substring_left_idx = 0
  substring_right_idx = 0
  substring_char = set()
  while right_idx <= len(string) - 1:
    if string[right_idx] not in substring_char:
      substring_char.add(string[right_idx])
      if right_idx - left_idx > substring_right_idx - substring_left_idx:
        substring_right_idx = right_idx
        substring_left_idx = left_idx
      right_idx += 1
    else:
      substring_char.remove(string[left_idx])
      left_idx += 1
  return string[substring_left_idx:substring_right_idx + 1]
````

````python
string = "clementisacap"
longestSubstringWithoutDuplication(string)
````

````python
string = ""
longestSubstringWithoutDuplication(string)
````

# Solution 2: move left idx when found duplicate to last seen of the element

move left_idx to max(last_seen, left_idx)

````python
def longestSubstringWithoutDuplication(string):
	left_idx = 0
	last_seen = {}
	substring_left_idx, substring_right_idx = 0, 0
	for right_idx in range(len(string)):
		char = string[right_idx]
		if char in last_seen:
			left_idx = max(left_idx, last_seen[char] + 1)
		if substring_right_idx - substring_left_idx < right_idx - left_idx:
			substring_right_idx, substring_left_idx = right_idx, left_idx
		last_seen[char] = right_idx
	return string[substring_left_idx:substring_right_idx + 1]
````

````python
string = "clementisacap"
longestSubstringWithoutDuplication(string)
````

````python
string = ""
longestSubstringWithoutDuplication(string)
````
