# Find Successor

Write a function that takes in a Binary Tree (where nodes have an additional pointer to their parent node) as well as a node contained in that tree and returns the given node's successor.

A node's successor is the next node to be visited (immediately after the given node) when traversing its tree using the in-order tree-traversal technique. A node has no successor if it's the last node to be visited in the in-order traversal.

If a node has no successor, your function should return `None` / `null`.

Each `BinaryTree` node has an integer `value`, a `parent` node, a `left` child node, and a `right` child node. Children nodes can either be `BinaryTree` nodes themselves or `None` / `null`.

# Solution 1: travere in-order from root

```python
# This is an input class. Do not edit.
class BinaryTree:
	def __init__(self, value, left=None, right=None, parent=None):
		self.value = value
		self.left = left
		self.right = right
		self.parent = parent

# Time O(n)
# Space O(n)
def findSuccessor(tree, node):
	in_order_traversal =  in_order_traverse(tree, node)
	
	for idx, current_node in enumerate(in_order_traversal):
		if node != current_node:
			continue
		else:
			if idx == len(in_order_traversal) - 1:
				return None
		return in_order_traversal[idx + 1]

def in_order_traverse(current_node, target_node, order = []):
	if current_node == None:
		return order

	in_order_traverse(current_node.left, target_node)
	order.append(current_node)
	in_order_traverse(current_node.right, target_node)

	return order
```

```python
tree = BinaryTree(1)
tree.left = BinaryTree(2, parent = tree)
tree.right = BinaryTree(3, parent = tree)
tree.left.left = BinaryTree(4, parent = tree.left)
tree.left.right = BinaryTree(5, parent = tree.left)
tree.left.left.left = BinaryTree(6, parent = tree.left.left)
node = tree.left.right
```

```python
findSuccessor(tree, node).value
```

# Solution 2: use parent properties of node

```python
# This is an input class. Do not edit.
class BinaryTree:
	def __init__(self, value, left=None, right=None, parent=None):
		self.value = value
		self.left = left
		self.right = right
		self.parent = parent

# Time O(h)
# Space O(1)
def findSuccessor(tree, node):
	if node.right != None:
		return get_left_most_child(node.right)

	return get_right_parent(node)

def get_left_most_child(node):
	while node.left != None:
		node = node.left
	return node

def get_right_parent(node):
	while node.parent != None:
		if node == node.parent.left:
			return node.parent
		node = node.parent

	return None
```

```python
tree = BinaryTree(1)
tree.left = BinaryTree(2, parent = tree)
tree.right = BinaryTree(3, parent = tree)
tree.left.left = BinaryTree(4, parent = tree.left)
tree.left.right = BinaryTree(5, parent = tree.left)
tree.left.left.left = BinaryTree(6, parent = tree.left.left)
node = tree.left.right
```

```python
findSuccessor(tree, node).value
```
