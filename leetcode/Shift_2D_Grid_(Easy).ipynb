# Shift 2D Grid

Given a 2D grid of size m x n and an integer k. You need to shift the grid k times.

In one shift operation:

- Element at grid[i][j] moves to grid[i][j + 1].
- Element at grid[i][n - 1] moves to grid[i + 1][0].
- Element at grid[m - 1][n - 1] moves to grid[0][0].

Return the 2D grid after applying shift operation k times.

image.png

Example 1:

```
Input: grid = [[1,2,3],[4,5,6],[7,8,9]], k = 1
Output: [[9,1,2],[3,4,5],[6,7,8]]
```



image.png

Example 2:

```
Input: grid = [[3,8,1,9],[19,7,2,5],[4,6,11,10],[12,0,21,13]], k = 4
Output: [[12,0,21,13],[3,8,1,9],[19,7,2,5],[4,6,11,10]]
```



Example 3:

```
Input: grid = [[1,2,3],[4,5,6],[7,8,9]], k = 9
Output: [[1,2,3],[4,5,6],[7,8,9]]
```



Constraints:

```
m == grid.length
n == grid[i].length
1 <= m <= 50
1 <= n <= 50
-1000 <= grid[i][j] <= 1000
0 <= k <= 100
```



````python
# Time O(n * m)
# Space O(n * m)

class Solution:
    def shiftGrid(self, grid, k):
        new_grid = [[0] * len(grid[0]) for _ in range(len(grid))]
        num_rows = len(grid)
        num_cols = len(grid[0])
        for row in range(num_rows):
            for col in range(num_cols):
                new_col = (col + k) % num_cols
                row_move = (col + k) // num_cols
                new_row = (row + row_move) % num_rows
                new_grid[new_row][new_col] = grid[row][col]
        return new_grid
````

````python
grid = [[1,2,3],[4,5,6],[7,8,9]]
k = 1
expected = [[9,1,2],[3,4,5],[6,7,8]]

output = Solution().shiftGrid(grid, k)
print(output)
assert output == expected
````

````python
grid = [[3,8,1,9],[19,7,2,5],[4,6,11,10],[12,0,21,13]]
k = 4
expected = [[12,0,21,13],[3,8,1,9],[19,7,2,5],[4,6,11,10]]

output = Solution().shiftGrid(grid, k)
print(output)
assert output == expected
````

````python
grid = [[1,2,3],[4,5,6],[7,8,9]]
k = 9
expected = [[1,2,3],[4,5,6],[7,8,9]]

output = Solution().shiftGrid(grid, k)
print(output)
assert output == expected
````
