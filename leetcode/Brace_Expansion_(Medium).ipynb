# Brace Expansion

You are given a string s representing a list of words. Each letter in the word has one or more options.

- If there is one option, the letter is represented as is.
- If there is more than one option, then curly braces delimit the options. For example, "{a,b,c}" represents options ["a", "b", "c"].

For example, if s = "a{b,c}", the first character is always 'a', but the second character can be 'b' or 'c'. The original list is ["ab", "ac"].

Return all words that can be formed in this manner, sorted in lexicographical order.

Example 1:

```
Input: s = "{a,b}c{d,e}f"
Output: ["acdf","acef","bcdf","bcef"]
```



Example 2:

```
Input: s = "abcd"
Output: ["abcd"]
```



Constraints:

```
1 <= s.length <= 50
s consists of curly brackets '{}', commas ',', and lowercase English letters.
s is guaranteed to be a valid input.
There are no nested curly brackets.
All characters inside a pair of consecutive opening and ending curly brackets are different.
```



````python
# Time = O(n^2 * m)
# Space = O(n^2 * m) (inc. output)
# Space = O(n) (exc. output)

# n = len of word
# m = number of options of each char

class Solution:
    def expand(self, s):
        split_string = s.replace('{', ' ').replace('}', ' ').strip().split(' ')
        split_string = [sorted(s.split(',')) for s in split_string]
        all_words = []
        expand_helper(split_string, 0, [], all_words)
        return all_words
        

def expand_helper(split_string, idx, current_word, all_words):
    if idx == len(split_string):
        all_words.append(''.join(current_word))
    else:
        for char in split_string[idx]:
            current_word.append(char)
            expand_helper(split_string, idx + 1, current_word, all_words)
            current_word.pop()
````

````python
s = "{a,b}c{d,e}f"
expected = ["acdf","acef","bcdf","bcef"]

output = Solution().expand(s)
print(output)
assert output == expected
````

````python
s = "abcd"
expected = ["abcd"]

output = Solution().expand(s)
print(output)
assert output == expected
````
