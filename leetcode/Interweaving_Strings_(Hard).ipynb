# Interweaving Strings

Write a function that takes in three strings and returns a boolean representing whether the third string can be formed by interweaving the first two strings.

To interweave strings means to merge them by alternating their letters without any specific pattern. For instance, the strings `"abc"` and `"123"` can be interwoven as `"a1b2c3"`, as `"abc123"`, and as `"ab1c23"` (this list is nonexhaustive).

Letters within a string must maintain their relative ordering in the interwoven string.

Sample Input

```
one = "algoexpert"
two = "your-dream-job"
three = "your-algodream-expertjob"
```



Sample Output

```
true
```



# Soltuion 1

````python
# Time O(2^(n+m)) : every recursion will have at most 2 dicision and it is n + m depths of recursion tree
# Space O(n+ m)
# n = len(one)
# m = len(two)
def interweavingStrings(one, two, three):
  if len(three) != len(one) + len(two):
    return False

  return are_interwoven(one, two, three, 0, 0)

def are_interwoven(one, two, three, idx1, idx2):
  idx3 = idx1 + idx2
  if idx3 == len(three):
    return True
  if idx1 <= len(one) - 1 and one[idx1] == three[idx3]:
    if are_interwoven(one, two, three, idx1 + 1, idx2):
      return True

  if idx2 <= len(two) - 1 and two[idx2] == three[idx3]:
    return are_interwoven(one, two, three, idx1, idx2 + 1)
  
  return False
````

````python
one = 'algoexpert'
three = 'your-algodream-expertjob'
two = 'your-dream-job'
interweavingStrings(one, two, three)
````

````python
one = 'aaa'
three = 'aaafaaa'
two = 'aaaf'
interweavingStrings(one, two, three)
````

# Solution 2: Dynamic Programming

````python
# Time O(nm)
# Space O(nm)

def interweavingStrings(one, two, three):
	if len(three) != len(one) + len(two):
		return False
	cache = [[None for _ in range(len(two) + 1)] for _ in range(len(one) + 1)]
	return are_interwoven(one, two, three, 0, 0, cache)

def are_interwoven(one, two, three, idx1, idx2, cache):
	if cache[idx1][idx2] != None:
		return cache[idx1][idx2]
	idx3 = idx1 + idx2
	if idx3 == len(three):
		cache[idx1][idx2] = True
		return True

	if idx1 <= len(one) - 1 and one[idx1] == three[idx3]:
		cache[idx1][idx2] = are_interwoven(one, two, three, idx1 + 1, idx2, cache)
		if cache[idx1][idx2]:
			return True

	if idx2 <= len(two) - 1 and two[idx2] == three[idx3]:
		cache[idx1][idx2] = are_interwoven(one, two, three, idx1, idx2 + 1, cache)
		return cache[idx1][idx2]

	cache[idx1][idx2] = False
	return False
````

````python
one = 'algoexpert'
three = 'your-algodream-expertjob'
two = 'your-dream-job'
interweavingStrings(one, two, three)
````

````python
one = 'aaa'
three = 'aaafaaa'
two = 'aaaf'
interweavingStrings(one, two, three)
````
