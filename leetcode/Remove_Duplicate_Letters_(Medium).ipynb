# Remove Duplicate Letters

Given a string s, remove duplicate letters so that every letter appears once and only once. You must make sure your result is the smallest in lexicographical order among all possible results.

Example 1:

```
Input: s = "bcabc"
Output: "abc"
```

Example 2:

```
Input: s = "cbacdcbc"
Output: "acdb"
```
Constraints:

```
1 <= s.length <= 104
s consists of lowercase English letters.
```


# Solution

````python
# Time = O(N)
# Space = O(1) including outout
# Sapce = O(1) excluding output
# space for last_idx_of_char, seen, stack is bounded by no. of unique char (26)

class Solution:
    def removeDuplicateLetters(self, s: str) -> str:
        stack = []
        seen = set()
        last_idx_of_char = {char: idx for idx, char in enumerate(s)}
        
        for idx, char in enumerate(s):
            if char in seen:
                continue
            while len(stack) > 0 and char < stack[-1] and idx < last_idx_of_char[stack[-1]]:
                seen.discard(stack.pop())
            seen.add(char)
            stack.append(char)
        return ''.join(stack)
````

````python
s = "bcabc"
expected = "abc"

output = Solution().removeDuplicateLetters(s)
print(output)
assert output == expected
````

````python
s = "cbacdcbc"
expected = "acdb"

output = Solution().removeDuplicateLetters(s)
print(output)
assert output == expected
````
