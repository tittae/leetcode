# Reverse Words In String

Write a function that takes in a string of words separated by one or more whitespaces and returns a string that has these words in reverse order. For example, given the string `"tim is great"`, your function should return `"great is tim"`.

For this problem, a word can contain special characters, punctuation, and numbers. The words in the string will be separated by one or more whitespaces, and the reversed string must contain the same whitespaces as the original string. For example, given the string `"whitespaces    4"` you would be expected to return `"4    whitespaces"`.

Note that you're not allowed to to use any built-in `split` or `reverse` methods/functions. However, you are allowed to use a built-in `join` method/function.

Also note that the input string isn't guaranteed to always contain words.

Sample Input

```
string = "AlgoExpert is the best!"
```



# Sample Output

```
"best! the is AlgoExpert"
```



# Solution 1

````python
# time = O(n)
# Altough, we have sting slice in for loop,
# string slice will happen no more than n time. Thus, it is like 2n in total.
# space = O(n)
# n = len(string)

def reverseWordsInString(string):
  words = []
  word_start_idx = 0
  for idx in range(len(string)):
    if string[idx] == " ":
      words.append(string[word_start_idx:idx])
      word_start_idx = idx
    elif string[word_start_idx] == " ":
      words.append(" ")
      word_start_idx = idx
  
  words.append(string[word_start_idx:])
  reversed_list(words)
  return ''.join(words)

def reversed_list(list):
  start, end = 0, len(list) - 1
  while start <= end:
    list[start], list[end] = list[end], list[start]
    start += 1
    end -= 1
````

````python
string = "AlgoExpert is the best!"
reverseWordsInString(string)
````

# Solution 2

````python
# Time = O(n)
# Space O(n)

def reverseWordsInString(string):
  chars = list(string)
  reversed_list(chars, 0, len(chars) - 1)
  start_idx = 0
  while start_idx < len(chars):
    end_idx = start_idx
    while end_idx < len(chars) and chars[end_idx] != ' ':
      end_idx += 1

    reversed_list(chars, start_idx, end_idx - 1)
    start_idx = end_idx + 1

  return ''.join(chars)


def reversed_list(list, start, end):
  while start <= end:
    list[start], list[end] = list[end], list[start]
    start += 1
    end -= 1
````

````python
string = "AlgoExpert is the best!"
reverseWordsInString(string)
````
