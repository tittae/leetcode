# Swapping Nodes in a Linked List

You are given the head of a linked list, and an integer k.

Return the head of the linked list after swapping the values of the kth node from the beginning and the kth node from the end (the list is 1-indexed).

image.png

Example 1:

```
Input: head = [1,2,3,4,5], k = 2
Output: [1,4,3,2,5]
```



Example 2:

```
Input: head = [7,9,6,6,7,8,3,0,9,5], k = 5
Output: [7,9,6,6,8,7,3,0,9,5]
```



Constraints:

```
The number of nodes in the list is n.
1 <= k <= n <= 105
0 <= Node.val <= 100
```



````python
# Time = O(N)
# Space = O(1)

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def __str__(self):
      node = self
      output_str = ''
      while node != None:
        output_str = output_str + str(node.val)
        if node.next != None:
          output_str = output_str + ' -> '
        node = node.next
      return output_str

class Solution:
    def swapNodes(self, head, k):
        left = head
        count = k - 1
        while count > 0:
            left = left.next
            count -= 1
        
        right = head
        tail = left
        while tail.next != None:
            tail = tail.next
            right = right.next
            
        left.val, right.val = right.val, left.val
        
        return head
````

````python
def create_link_list(array):
  head = ListNode(array[0])
  node = head
  for i in range(1, len(array)):
    node.next = ListNode(array[i])
    node = node.next
  return head
````

````python
head = create_link_list([7,9,6,6,7,8,3,0,9,5])
k = 5
expected = create_link_list([7,9,6,6,8,7,3,0,9,5])

output = Solution().swapNodes(head, k)
print(output)
assert output.__str__() == expected.__str__()
````

````python
head = create_link_list([1,2,3,4,5])
k = 2
expected = create_link_list([1,4,3,2,5])

output = Solution().swapNodes(head, k)
print(output)
assert output.__str__() == expected.__str__()
````
