# Balance Index

Write a function that takes in a non-empty array of integers and returns its balance index.

An array `a`'s balance index is an index i such that `a[0] + a[1] + ... + a[i - 1]` is equal to `a[i + 1] + a[i + 2] + ... + a[lastIdx]`.

In order for `0` to be an array's balance index, `a[1] + a[2] + ... + a[lastIdx]` must be equal to `0`.

Similarly, in order for the last index of an array to be the array's balance index, `a[0] + a[1] + ... + a[lastIdx - 1]` must be equal to `0`.

If there is no balance index in the array, your function should return `-1`; if there are multiple possible answers, your function can return any of them.

Sample Input

```
array = [0, 9, -8, 2, 7, 1, 11, -2, 1]
```



Sample Output

```
5 // 0 + 9 + -8 + 2 + 7 == 11 + -2 + 1
```



# Solution 1

left traverse and then right traverse

````python
# Time O(n)
# Space O(n)

def balanceIndex(array):
  left_sum = [None for _ in range(len(array))]
  left_sum[0] = array[0]
  for left_idx in range(1, len(array)):
    left_sum[left_idx] = left_sum[left_idx - 1] + array[left_idx]

  right_sum = 0
  for right_idx in range(len(array) - 1, -1, -1):
    right_sum += array[right_idx]
    if right_sum == left_sum[right_idx]:
      return right_idx

  return -1
````

````python
array = [0, 9, -8, 2, 7, 1, 11, -2, 1]
balanceIndex(array)
````

````python
array = [0, 9, -8, 2, 7, 1, 11, -2, 1, 1, -22]
balanceIndex(array)
````

````python
array = [-22, 0, 9, -8, 2, 7, 1, 11, -2, 1, 1, 30]
balanceIndex(array)
````

# Solution 2

only left traverse

````python
def balanceIndex(array):
  left_sum = 0
  array_sum = sum(array)
  for i in range(len(array)):
    if array_sum == 2*left_sum + array[i]:
      return i
    left_sum += array[i]

  return -1 
````

````python
array = [0, 9, -8, 2, 7, 1, 11, -2, 1]
balanceIndex(array)
````

````python
array = [0, 9, -8, 2, 7, 1, 11, -2, 1, 1, -22]
balanceIndex(array)
````

````python
array = [-22, 0, 9, -8, 2, 7, 1, 11, -2, 1, 1, 30]
balanceIndex(array)
````
