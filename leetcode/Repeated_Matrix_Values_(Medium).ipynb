# Repeated Matrix Values

Write a function that takes in a non-empty two-dimensional integer array (a matrix) of potentially unequal height and width and returns a list of the elements that appear at least once in each row and each column of the matrix.

Sample Input

```
matrix = [
  [1, 3, 7, 4, 5],
  [2, 5, 9, 3, 3],
  [1, 8, 5, 3, 5],
  [5, 0, 3, 5, 6],
  [3, 8, 3, 5, 6],
  [1, 0, 3, 0, 5],
]
```



Sample Output

```
[3, 5] // The numbers could be ordered differently.
```



# Solution

````python
# Time O(wh)
# Sapce O(min(w, h))

def repeatedMatrixValues(matrix):
  previous_valid_elements = get_initial_element(matrix)
  for row in matrix:
    valid_elements = set()
    for num in row:
      if num in previous_valid_elements:
        valid_elements.add(num)
    previous_valid_elements = valid_elements

  for col_idx in range(len(matrix[0])):
    valid_elements = set()
    for row_idx in range(len(matrix)):
      num = matrix[row_idx][col_idx]
      if num in previous_valid_elements:
        valid_elements.add(num)
    previous_valid_elements = valid_elements

  return list(previous_valid_elements)
    

def get_initial_element(matrix):
  width = len(matrix[0])
  height = len(matrix)
  elements = set()
  if width < height:
    for num in matrix[0]:
      elements.add(num)
  else:
    for idx in range(len(matrix)):
      num = matrix[idx][0]
      elements.add(num)
  return elements
````

````python
matrix = [
    [1, 3, 7, 4, 5],
    [2, 5, 9, 3, 3],
    [1, 8, 5, 3, 5],
    [5, 0, 3, 5, 6],
    [3, 8, 3, 5, 6],
    [1, 0, 3, 0, 5]
  ]
repeatedMatrixValues(matrix)
````

````python
matrix = [[]]
repeatedMatrixValues(matrix)
````
