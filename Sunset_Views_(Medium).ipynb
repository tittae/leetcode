# Sunset Views

Given an array of buildings and a direction that all of the buildings face, return an array of the indices of the buildings that can see the sunset.

A building can see the sunset if it's strictly taller than all of the buildings that come after it in the direction that it faces.

The input array named `buildings` contains positive, non-zero integers representing the heights of the buildings. A building at index `i` thus has a height denoted by `buildings[i]`. All of the buildings face the same direction, and this direction is either east or west, denoted by the input string named `direction`, which will always be equal to either `"EAST"` or `"WEST"`. In relation to the input array, you can interpret these directions as right for east and left for west.

Important note: the indices in the ouput array should be sorted in ascending order.

Sample Input #1

```
buildings = [3, 5, 4, 4, 3, 1, 3, 2]
direction = "EAST"
```



Sample Output #1

```
[1, 3, 6, 7]

```



Sample Input #2

```
buildings = [3, 5, 4, 4, 3, 1, 3, 2]
direction = "WEST"
```



Sample Output #2

```
[0, 1]

// The buildings are the same as in the first sample
// input, but their direction is reversed.
```



````python
# Time O(n)
# Space O(n)

def sunsetViews(buildings, direction):
  tallest_height = 0
  buildings_with_sunset_views = []
  if direction == 'EAST':
    for idx in reversed(range(len(buildings))):
      if buildings[idx] > tallest_height:
        tallest_height = buildings[idx]
        buildings_with_sunset_views.append(idx)
    buildings_with_sunset_views = list(reversed(buildings_with_sunset_views))

  elif direction == 'WEST':
    for idx in range(len(buildings)):
      if buildings[idx] > tallest_height:
        tallest_height = buildings[idx]
        buildings_with_sunset_views.append(idx)
  
  return buildings_with_sunset_views
````

````python
buildings = [3, 5, 4, 4, 3, 1, 3, 2]
direction = "EAST"
sunsetViews(buildings, direction)
````

````python
buildings = [3, 5, 4, 4, 3, 1, 3, 2]
direction = "WEST"
sunsetViews(buildings, direction)
````

# Solution 2

Same as solution 1 but refactoring

````python
def sunsetViews(buildings, direction):
  tallest_height = 0
  buildings_with_sunset_views = []
  if direction == 'EAST':
    idx = len(buildings) - 1
    step = -1
  elif direction == 'WEST':
    idx = 0
    step = 1 

  while idx >= 0 and idx <= len(buildings) - 1:
    if buildings[idx] > tallest_height:
      buildings_with_sunset_views.append(idx)
      tallest_height = buildings[idx]
    idx += step

  if direction == 'EAST':
    buildings_with_sunset_views = list(reversed(buildings_with_sunset_views))
  
  return buildings_with_sunset_views
````

````python
def sunsetViews(buildings, direction):
  buildings_with_sunset_views = []
  if direction == 'WEST':
    idx = len(buildings) - 1
    step = -1
  elif direction == 'EAST':
    idx = 0
    step = 1 
  
  while idx >= 0 and idx <= len(buildings) - 1:
    while len(buildings_with_sunset_views) > 0 and buildings[buildings_with_sunset_views[-1]] <= buildings[idx]:
      buildings_with_sunset_views.pop()
    
    buildings_with_sunset_views.append(idx)
    idx += step
  
  if direction == 'WEST':
    buildings_with_sunset_views = list(reversed(buildings_with_sunset_views))
  
  return buildings_with_sunset_views
````

````python
buildings = [3, 5, 4, 4, 3, 1, 3, 2]
direction = "EAST"
sunsetViews(buildings, direction)
````

````python
buildings = [3, 5, 4, 4, 3, 1, 3, 2]
direction = "WEST"
sunsetViews(buildings, direction)
````
