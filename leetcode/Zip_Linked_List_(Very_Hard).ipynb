# Zip Linked List

You're given the head of a Singly Linked List of arbitrary length `k`. Write a function that zips the Linked List in place (i.e., doesn't create a brand new list) and returns its head.

A Linked List is zipped if its nodes are in the following order, where `k` is the length of the Linked List:


```
1st node -> kth node -> 2nd node -> (k - 1)th node -> 3rd node -> (k - 2)th node -> ...
```



Each `LinkedList` node has an integer `value` as well as a `next` node pointing to the next node in the list or to `None` / `null` if it's the tail of the list.

You can assume that the input Linked List will always have at least one node; in other words, the head will never be `None` / `null`.

Sample Input

```
linkedList = 1 -> 2 -> 3 -> 4 -> 5 -> 6 // the head node with value 1 
```



Sample Output

```
1 -> 6 -> 2 -> 5 -> 3 -> 4 // the head node with value 1
```



````python
# This is an input class. Do not edit.
class LinkedList:
  def __init__(self, value):
    self.value = value
    self.next = None
          
  def __str__(self):
    node = self
    output_str = ''
    while node != None:
      output_str = output_str + str(node.value)
      if node.next != None:
        output_str = output_str + ' -> '
      node = node.next
    return output_str
````

````python
# Time O(n)
# Space O(1)
def zipLinkedList(linkedList):
  if linkedList.next == None:
    return linkedList
  first_head = linkedList
  second_head = split_linkedlist(linkedList)
  reversed_second_head = reverse_linkedlist(second_head)
  head = alternate_linkedlist(first_head, reversed_second_head)
  return head

# Time O(n)
# Space O(1)
def split_linkedlist(linkedList):
  node = linkedList
  idx = 0
  while node != None:
    node = node.next
    idx += 1

  idx = round(idx/2)
  second_head = linkedList
  first_tail = None
  for i in range(idx):
    first_tail = second_head
    second_head = second_head.next

  first_tail.next = None
  return second_head

# Time O(n)
# Space O(1)
def reverse_linkedlist(head):
	p1 = None
	p2 = head
	while p2 != None:
		p3 = p2.next
		p2.next = p1
		p1 = p2
		p2 = p3

	return p1

# Time O(n)
# Space O(1)
def alternate_linkedlist(first_head, second_head):
  p1 = first_head.next
  p2 = second_head
  head = first_head
  tail = first_head
  while p1 != None or p2 != None:
    if p2 != None:
      tail.next = p2
      tail = p2
      p2 = p2.next

    if p1 != None:
      tail.next = p1
      tail = p1
      p1 = p1.next

  return first_head
````

````python
def create_link_list(array):
  head = LinkedList(array[0])
  node = head
  for i in range(1, len(array)):
    node.next = LinkedList(array[i])
    node = node.next
  return head
````

````python
head = create_link_list(range(1, 7))
print(head)
````

````python
head = zipLinkedList(head)
print(head)
````
