# Number Of Ways To Traverse Graph

You're given two positive integers representing the width and height of a grid-shaped, rectangular graph. Write a function that returns the number of ways to reach the bottom right corner of the graph when starting at the top left corner. Each move you take must either go down or right. In other words, you can never move up or left in the graph.

For example, given the graph illustrated below, with width = 2 and height = 3, there are three ways to reach the bottom right corner when starting at the top left corner:

```
 _ _
|_|_|
|_|_|
|_|_|
```
1. Down, Down, Right
2. Right, Down, Down
3. Down, Right, Down

Note: you may assume that width * height >= 2. In other words, the graph will never be a 1x1 grid.




Sample Input

```
width = 4
height = 3
```



Sample Output

```
10
```



````python
# Time O(wh)
# Space O(min(w, h))

def numberOfWaysToTraverseGraph(width, height):
	if width < height:
		short_lenght, long_lenght = width, height
	else:
		short_lenght, long_lenght = height, width

	previous_row = [1] * (short_lenght + 1)
	current_row = [1] * (short_lenght + 1)

	for j in range(2, long_lenght + 1):
		for i in range(2, short_lenght + 1):
			current_row[i] = current_row[i - 1] + previous_row[i]
		
		previous_row = current_row[:]

	return current_row[-1]
````

````python
width = 4
height = 3
print(numberOfWaysToTraverseGraph(width, height))
````

````python
# Time O(w + h)
# Space O(1)

def numberOfWaysToTraverseGraph(width, height):
	n = (width - 1) + (height - 1)
	r = (width - 1)

	n_frac = fractorial(n)
	r_frac = fractorial(r)
	n_minus_r_frac = fractorial(n - r)
	
	permutation = n_frac / (r_frac * n_minus_r_frac)

	return permutation


def fractorial(n):
	frac = 1
	for i in range(1, n + 1):
		frac *= i
	return frac
````

````python
width = 4
height = 3
print(numberOfWaysToTraverseGraph(width, height))
````
