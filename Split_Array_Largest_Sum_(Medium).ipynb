# Split Array Largest Sum

Given an array nums which consists of non-negative integers and an integer m, you can split the array into m non-empty continuous subarrays.

Write an algorithm to minimize the largest sum among these m subarrays.

Example 1:

```
Input: nums = [7,2,5,10,8], m = 2
Output: 18
Explanation:
There are four ways to split nums into two subarrays.
The best way is to split it into [7,2,5] and [10,8],
where the largest sum among the two subarrays is only 18.
```



Example 2:

```
Input: nums = [1,2,3,4,5], m = 2
Output: 9
```



Example 3:

```
Input: nums = [1,4,4], m = 3
Output: 4
```



Constraints:

```
1 <= nums.length <= 1000
0 <= nums[i] <= 106
1 <= m <= min(50, nums.length)
```



# Solution 1:

- Binary Search

````python
# time = O(n log(s))
# space = O(1)

# s = sum(nums) - max(nums)
# n = len(nums)

class Solution:
    def splitArray(self, nums, m):
        left = max(nums)
        right = sum(nums)
        while left <= right:
            mid = (left + right) // 2
            if get_min_required_subarray(nums, mid) <= m:
                right = mid - 1
                min_largest_split_sum = mid
            else:
                left = mid + 1
        return min_largest_split_sum

def get_min_required_subarray(nums, max_sum_allowed):
    required_split = 0
    current_sum = 0
    for num in nums:
        if current_sum + num > max_sum_allowed:
            current_sum = num
            required_split += 1
        else:
            current_sum += num
    return required_split + 1
````

````python
nums = [7,2,5,10,8]
m = 2
expected = 18

output = Solution().splitArray(nums, m)
print(output)
assert output == expected
````

````python
nums = [1,2,3,4,5]
m = 2
expected = 9

output = Solution().splitArray(nums, m)
print(output)
assert output == expected
````

````python
nums = [1,2,3,4,5]
m = 3
expected = 6

output = Solution().splitArray(nums, m)
print(output)
assert output == expected
````

````python
nums = [1,4,4]
m = 3
expected = 4

output = Solution().splitArray(nums, m)
print(output)
assert output == expected
````
