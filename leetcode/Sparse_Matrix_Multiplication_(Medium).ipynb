# Sparse Matrix Multiplication

Given two sparse matrices mat1 of size m x k and mat2 of size k x n, return the result of mat1 x mat2. You may assume that multiplication is always possible.

 

image.png

Example 1:

```
Input: mat1 = [[1,0,0],[-1,0,3]], mat2 = [[7,0,0],[0,0,0],[0,0,1]]
Output: [[7,0,0],[-7,0,3]]
```



Example 2:

```
Input: mat1 = [[0]], mat2 = [[0]]
Output: [[0]]
```



Constraints:

```
m == mat1.length
k == mat1[i].length == mat2.length
n == mat2[i].length
1 <= m, n, k <= 100
-100 <= mat1[i][j], mat2[i][j] <= 100
```



````python
# Time = O(pqr)
# Space = O(pq + qr)
# where size of mat1 = p * q, mat2 = q * r


class Solution:
    def multiply(self, mat1, mat2):
        mat1_nonzero_cells = get_nonzero_cells(mat1)
        mat2_nonzero_cells = get_nonzero_cells(mat2)
        
        mat = [[0 for _ in range(len(mat2[0]))] for _ in range(len(mat1))]
        
        for i, k in mat1_nonzero_cells.keys():
            for j in range(len(mat2[0])):
                if (k, j) in mat2_nonzero_cells:
                    mat[i][j] += mat1_nonzero_cells[(i, k)] * mat2_nonzero_cells[(k, j)]
        return mat
    
    
def get_nonzero_cells(matrix):
    nonzero_cells = {}
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if matrix[i][j] != 0:
                nonzero_cells[(i, j)] = matrix[i][j]
    return nonzero_cells
````

````python
mat1 = [[1,0,0],[-1,0,3]]
mat2 = [[7,0,0],[0,0,0],[0,0,1]]
expected = [[7,0,0],[-7,0,3]]

output = Solution().multiply(mat1, mat2)
print(output)
assert output == expected
````

````python
mat1 = [[0]]
mat2 = [[0]]
expected = [[0]]

output = Solution().multiply(mat1, mat2)
print(output)
assert output == expected
````
