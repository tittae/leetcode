# All Kinds Of Node Depths

The distance between a node in a Binary Tree and the tree's root is called the node's depth.

Write a function that takes in a Binary Tree and returns the sum of all of its subtrees' nodes' depths.

Each `BinaryTree` node has an integer `value`, a `left` child node, and a `right` child node. Children nodes can either be `BinaryTree` nodes themselves or `None` / `null`.

Sample Input

```
tree =    1
       /     \
      2       3
    /   \   /   \
   4     5 6     7
 /   \
8     9
```



Sample Output

```
26
// The sum of the root tree's node depths is 16.
// The sum of the tree rooted at 2's node depths is 6.
// The sum of the tree rooted at 3's node depths is 2.
// The sum of the tree rooted at 4's node depths is 2.
// Summing all of these sums yields 26.
```



# Solution 1

- Literal implementation

````python
# Time : O(n)
# Space : O(h)

def allKindsOfNodeDepths(root):
	return get_tree_info(root).sum_depths_all_kinds

def get_tree_info(root):
	if root == None:
		return tree_info(0, 0, 0)
	left_tree_info = get_tree_info(root.left)
	right_tree_info = get_tree_info(root.right)
	sum_depths = left_tree_info.sum_depths + right_tree_info.sum_depths + left_tree_info.num_nodes + right_tree_info.num_nodes
	num_nodes = left_tree_info.num_nodes + right_tree_info.num_nodes + 1
	sum_depths_all_kinds = sum_depths + left_tree_info.sum_depths_all_kinds + right_tree_info.sum_depths_all_kinds
	
	return tree_info(num_nodes, sum_depths, sum_depths_all_kinds)

class tree_info:
	def __init__(self, num_nodes, sum_depths, sum_depths_all_kinds):
		self.num_nodes = num_nodes
		self.sum_depths = sum_depths
		self.sum_depths_all_kinds = sum_depths_all_kinds

# This is the class of the input binary tree.
class BinaryTree:
	def __init__(self, value):
		self.value = value
		self.left = None
		self.right = None
````

````python
root = BinaryTree(1)
root.left = BinaryTree(2)
root.right = BinaryTree(3)
root.left.left = BinaryTree(4)
root.left.right = BinaryTree(5)
root.right.left = BinaryTree(6)
root.right.right = BinaryTree(7)
root.left.left.left = BinaryTree(8)
root.left.left.right = BinaryTree(9)
````

````python
allKindsOfNodeDepths(root)
````

# Solution 2

- Apply a formula

````python
# Time : O(n)
# Space : O(h)

def allKindsOfNodeDepths(root, depth = 0):
	if root == None:
		return 0
	depth_sum = (depth * (depth + 1)) / 2
	return depth_sum + allKindsOfNodeDepths(root.left, depth + 1) + allKindsOfNodeDepths(root.right, depth + 1)

# This is the class of the input binary tree.
class BinaryTree:
	def __init__(self, value):
		self.value = value
		self.left = None
		self.right = None
````

````python
root = BinaryTree(1)
root.left = BinaryTree(2)
root.right = BinaryTree(3)
root.left.left = BinaryTree(4)
root.left.right = BinaryTree(5)
root.right.left = BinaryTree(6)
root.right.right = BinaryTree(7)
root.left.left.left = BinaryTree(8)
root.left.left.right = BinaryTree(9)
````

````python
allKindsOfNodeDepths(root)
````
